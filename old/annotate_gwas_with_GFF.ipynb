```python
# from https://stackoverflow.com/questions/268272/getting-key-with-maximum-value-in-dictionary
def keywithmaxval(d):
     """ a) create a list of the dict's keys and values; 
         b) return the key with the max value"""  
     v=list(d.values())
     k=list(d.keys())
     return k[v.index(max(v))]
```

##  Pick triplet using ref_position and feature_position_mod3
    mod3 == 0, third base, triplet (x-2):x
    mod3 ==1, first base,   triplet x:(x+2)
    mod3 ==2, 2nd base,     triplet, (x-1):(x+1)
    
    
ORF1ab join (266..13468, 13468..21555). <br>
(13468-266+1)/3 = 4401 <br>
(21555-13468+1/3 = 2696 <br> 
In Python string, the slippage region is 13467 .. 21554 in string.  <br>
Visual check of the csv output shows that at least 5 mutations coveres the C-13467 positions

```python
def get_codon_coord( in_ref_pos, in_mod3 ):
    if ( in_mod3 == 0): 
        #third base
        return( (in_ref_pos - 2, in_ref_pos) )
    elif ( in_mod3 == 1): 
        #first base
        return( (in_ref_pos, in_ref_pos +2 ) )
    elif ( in_mod3 == 2) : 
        #second base
        return ( (in_ref_pos - 1, in_ref_pos + 1) )
    else : #error 
        return( (-1, -1) )

print(get_codon_coord(2, 0) )
print(get_codon_coord(0, 1) )
print(get_codon_coord(1, 2) )
```

## Generate mutant triplet based on reference triplet

ref_triplet <br>
mutant_nt <br> 
mod3 <br> 

mod3 == 0, third base, 2-nd <br> 
mod3 == 1, first base, 0-th <br>
mod3 == 2, second base, 1-th  <br> 

```python
def mutate_refTriplet_by_mod3 (ref_triplet, mutant_nt, mod3):
    triplet = ref_triplet[:]
    if ( mod3==0 ):
        triplet = triplet[0:2] + mutant_nt
    elif ( mod3==1 ):
        triplet = mutant_nt + triplet[1] + triplet[2]
    elif ( mod3==2): 
        triplet = triplet[0] + mutant_nt + triplet[2]
    return( triplet )

print( mutate_refTriplet_by_mod3 ('ATG', 'X', 0))
print( mutate_refTriplet_by_mod3 ('ATG', 'X', 1))
print( mutate_refTriplet_by_mod3 ('ATG', 'X', 2))
```

```python
# convert DNA positions into codon positions
def get_ref_triplet_position_by_mod3 (ref_dna_position):
    mod3 = ref_dna_position % 3
    triplet_position = ref_dna_position // 3
    if ( mod3 == 0 ):
        return triplet_position
    elif ( mod3 == 1 ) or (mod3 == 2):
        return (triplet_position + 1)
    else:
        return float("NaN")
   
print( get_ref_triplet_position_by_mod3(9) )
print( get_ref_triplet_position_by_mod3(10) )
print( get_ref_triplet_position_by_mod3(11) )
```
